# Generate HCAHPS dataframe

```python
import pandas as pd
import numpy as np
import warnings
from IPython.utils import io
import sys
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

main_dir = '/Users/kenlocey/Desktop/Rush/CMS_HospitalArchives/'
```

```python
hcp_df = pd.read_pickle(main_dir + 'HCAHPS/CombinedFiles_HCAHPS/' + 'Facility.pkl')

hcp_df.head()
```

```python

hcp_df = hcp_df.drop(labels=['Address', 'City', 'County', 'End Date',
                            'Facility Name', 'Phone Number', 'Footnote', 
                             'Start Date', 'State', 'ZIP Code',
                            ], axis=1)

for i in list(hcp_df):
    if 'ootnote' in i:
        hcp_df.drop(labels=[i], axis=1, inplace=True)
print(hcp_df.shape)
print(list(hcp_df))

hcp_df = hcp_df[~hcp_df['HCAHPS Answer Description'].isin([np.nan, float("NaN")])]
labs = hcp_df['HCAHPS Answer Description'].tolist()

hcp_df.head()
```

```python

labs2 = []
for l in labs:
    
    if "\x93" in l:
        l = l.replace("\x93", '')
    if "\x94" in l:
        l = l.replace("\x94", '')
    if "  " in l:
        l = l.replace("  ", ' ')

    labs2.append(l)

    
labs = sorted(list(set(labs2)))

print(len(labs), 'labels')
for l in labs:
    print(l)
```

```python
hcp_df['HCAHPS Answer Description'] = labs2
```

```python
hcp_df1 = hcp_df.filter(items=['Facility ID',
                               'file_month',
                               'file_year',
                               'HCAHPS Answer Description',
                               'HCAHPS Linear Mean Value',
                               ], axis=1)
print(hcp_df1.shape)
hcp_df1 = hcp_df1[hcp_df1['HCAHPS Linear Mean Value'] != 'Not Applicable']
hcp_df1['HCAHPS Answer Description'].fillna('None', inplace=True)
hcp_df1 = hcp_df1[hcp_df1['HCAHPS Answer Description'] != 'None']
print(hcp_df1.shape)
hcp_df1.drop_duplicates(inplace=True)
print(hcp_df1.shape)


labels = hcp_df1['HCAHPS Answer Description'].unique()
main_df = pd.DataFrame(columns=['Facility ID', 'file_month', 'file_year'])

for i, l in enumerate(labels):
    
    tdf = hcp_df1[hcp_df1['HCAHPS Answer Description'] == l]
    hcp_df12 = pd.DataFrame(columns=['Facility ID', 'file_month', 'file_year'])

    for n in list(tdf):
        if n == 'HCAHPS Answer Description' or n in ['Facility ID', 'file_month', 'file_year']:
                continue

        else:
            tdf[n] = pd.to_numeric(tdf[n], errors='coerce')
            tdf.rename(columns={n: l}, inplace=True)
        
        tdf.drop(labels=['HCAHPS Answer Description'], axis=1, inplace=True)
        
        hcp_df12 = hcp_df12.merge(tdf, on=['Facility ID', 'file_month', 
                                          'file_year'], how='outer')
    
    main_df = main_df.merge(hcp_df12, on=['Facility ID', 'file_month', 
                                          'file_year'], how='outer')

hcp_df1 = main_df.copy(deep=True)
del main_df, hcp_df12
print(hcp_df1.shape)
hcp_df1.dropna(how='all', axis=1, inplace=True)
print(hcp_df1.shape)
hcp_df1.head()
```

```python
hcp_df2 = hcp_df.filter(items=['Facility ID',
                               'file_month',
                               'file_year',
                               'HCAHPS Answer Description',
                               'Patient Survey Star Rating',
                               ], axis=1)
print(hcp_df2.shape)
hcp_df2 = hcp_df2[hcp_df2['Patient Survey Star Rating'] != 'Not Applicable']
hcp_df2['HCAHPS Answer Description'].fillna('None', inplace=True)
hcp_df2 = hcp_df2[hcp_df2['HCAHPS Answer Description'] != 'None']
print(hcp_df2.shape)
hcp_df2.drop_duplicates(inplace=True)
print(hcp_df2.shape)


labels = hcp_df2['HCAHPS Answer Description'].unique()
main_df = pd.DataFrame(columns=['Facility ID', 'file_month', 'file_year'])

for i, l in enumerate(labels):
    
    tdf = hcp_df2[hcp_df2['HCAHPS Answer Description'] == l]
    hcp_df22 = pd.DataFrame(columns=['Facility ID', 'file_month', 'file_year'])

    for n in list(tdf):
        if n == 'HCAHPS Answer Description' or n in ['Facility ID', 'file_month', 'file_year']:
                continue

        else:
            tdf[n] = pd.to_numeric(tdf[n], errors='coerce')
            tdf.rename(columns={n: l}, inplace=True)
        
        tdf.drop(labels=['HCAHPS Answer Description'], axis=1, inplace=True)
        
        hcp_df22 = hcp_df22.merge(tdf, on=['Facility ID', 'file_month', 
                                          'file_year'], how='outer')
    
    main_df = main_df.merge(hcp_df22, on=['Facility ID', 'file_month', 
                                          'file_year'], how='outer')

hcp_df2 = main_df.copy(deep=True)
del main_df, hcp_df22
print(hcp_df2.shape)
hcp_df2.dropna(how='all', axis=1, inplace=True)
print(hcp_df2.shape)
hcp_df2.head()
```

```python
hcp_df3 = hcp_df.filter(items=['Facility ID',
                               'file_month',
                               'file_year',
                               'HCAHPS Answer Description',
                               'HCAHPS Answer Percent',
                               ], axis=1)
print(hcp_df3.shape)
hcp_df3 = hcp_df3[hcp_df3['HCAHPS Answer Percent'] != 'Not Applicable']
hcp_df3['HCAHPS Answer Description'].fillna('None', inplace=True)
hcp_df3 = hcp_df3[hcp_df3['HCAHPS Answer Description'] != 'None']
print(hcp_df3.shape)
hcp_df3.drop_duplicates(inplace=True)
print(hcp_df3.shape)


labels = hcp_df3['HCAHPS Answer Description'].unique()
main_df = pd.DataFrame(columns=['Facility ID', 'file_month', 'file_year'])

for i, l in enumerate(labels):
    
    tdf = hcp_df3[hcp_df3['HCAHPS Answer Description'] == l]
    hcp_df32 = pd.DataFrame(columns=['Facility ID', 'file_month', 'file_year'])

    for n in list(tdf):
        if n == 'HCAHPS Answer Description' or n in ['Facility ID', 'file_month', 'file_year']:
                continue

        else:
            tdf[n] = pd.to_numeric(tdf[n], errors='coerce')
            tdf.rename(columns={n: l}, inplace=True)
        
        tdf.drop(labels=['HCAHPS Answer Description'], axis=1, inplace=True)
        
        hcp_df32 = hcp_df32.merge(tdf, on=['Facility ID', 'file_month', 
                                          'file_year'], how='outer')
    
    main_df = main_df.merge(hcp_df32, on=['Facility ID', 'file_month', 
                                          'file_year'], how='outer')

hcp_df3 = main_df.copy(deep=True)
del main_df, hcp_df32
print(hcp_df3.shape)
hcp_df3.dropna(how='all', axis=1, inplace=True)
print(hcp_df3.shape)
hcp_df3.head()
```

```python
hcp_df4 = hcp_df.filter(items=['Facility ID',
                               'file_month',
                               'file_year',
                               'Number of Completed Surveys',
                               'Survey Response Rate',
                               'Survey Response Rate Percent',
                               ], axis=1)
print(hcp_df4.shape)
hcp_df4.drop_duplicates(inplace=True)
hcp_df4.dropna(how='all', axis=1, inplace=True)
print(hcp_df4.shape)
hcp_df4.head()
```

```python
main_df = hcp_df1.merge(hcp_df2, how='outer', on=['Facility ID', 'file_month', 'file_year'])
main_df = main_df.merge(hcp_df3, how='outer', on=['Facility ID', 'file_month', 'file_year'])
main_df = main_df.merge(hcp_df4, how='outer', on=['Facility ID', 'file_month', 'file_year'])

for i, n in enumerate(list(main_df)):
    if n in ['Facility ID', 'file_month', 'file_year']:
        continue
            
    main_df.rename(columns={n: 'HCAHPS: ' + n}, inplace=True)

print(main_df.shape)
main_df.head()
```

```python
initial = hcp_df.shape[0] * hcp_df.shape[1]
final = main_df.shape[0] * main_df.shape[1]

p_d = round(100 * (initial - final)/initial, 2)

print(str(p_d)+'% decrease in size of HCAHPS data by use of efficient format.')
del hcp_df
```

```python
main_df.to_pickle('dataframe_data/hcahps_df.pkl', protocol=5)
```
