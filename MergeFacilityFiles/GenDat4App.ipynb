# Generate a pkl file for populating the Hospital Compare App with data

```python
import pandas as pd
import numpy as np
import warnings
from IPython.utils import io
import sys
import csv

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

main_dir = '~/GitHub/hospitals-data-archive/'
```

```python
main_df = pd.read_pickle(main_dir + 'MergeFacilityFiles/dataframe_data/all_hospitals_df.pkl.gz', compression='gzip')
main_df.head()
```

```python
ls = [('Name and Num', 'Name and Num'), ('Facility ID', 'Facility ID'), ('file_year', 'file_year'), ('file_month', 'file_month'),
      ('General Info', 'Facility Name'), ('General Info', 'Emergency Services'),
      ('General Info', 'Hospital Ownership'), ('General Info', 'Hospital Type'),
      ('General Info', 'State'),
     
     ]
        
df1 = main_df.filter(items=ls, axis=1)
df1 = df1.set_index(('Facility ID', 'Facility ID'))
df1.columns = df1.columns.droplevel(0)
df1 = df1.reset_index()
df1.rename(columns={('Facility ID', 'Facility ID'): 'Facility ID'}, inplace=True)
df1 = df1.filter(items=['Name and Num', 'Facility ID', 'Facility Name',
                        'Emergency Services', 'Hospital Ownership', 'Hospital Type', 'State'
                       ], axis=1)

print(df1.shape)
df1.drop_duplicates(inplace=True)
print(df1.shape)
df1.head()
```

```python
df = pd.read_pickle(main_dir + 'MergeFacilityFiles/dataframe_data/Files_For_App/GenDat4App_p4.pkl')
df.head()
```

```python
df2 = df.set_index(('Curated Name and Num', 'Curated Name and Num', 'Curated Name and Num', 'Curated Name and Num'))
df2.columns = df2.columns.droplevel([0,1,2])
df2 = df2.reset_index()
df2.rename(columns={'Hospital Provider Number (PRVDR_NUM)': 'Facility ID',
                    'Total Facility (S3_1_C2_27)': 'Beds',
                   }, inplace=True)
df2 = df2.filter(items=['Facility ID', 'Beds', 'Lat', 'Lon',
                       ], axis=1)


df2['Beds'] = df2.groupby(['Facility ID'])['Beds'].transform('mean')
df2['Beds'] = np.round(df2['Beds'], 0)

print(df2.shape)
df2.drop_duplicates(inplace=True)
print(df2.shape)
df2.head()
```

```python
df = df1.merge(df2, how='outer', on=['Facility ID'])

print(df.shape)
df.drop_duplicates(inplace=True)
print(df.shape)
df.head()
```

```python
df.to_pickle(main_dir + 'MergeFacilityFiles/dataframe_data/Files_For_App/GenDat4App.pkl', protocol=4)
```
