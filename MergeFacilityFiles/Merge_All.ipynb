# Merge all Hospital Archives dataframes

```python
import pandas as pd
import numpy as np
import warnings
from IPython.utils import io
import sys
import csv

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

main_dir = '~/GitHub/hospitals-data-archive/'
```

```python
main_df = pd.read_pickle(main_dir + 'MergeFacilityFiles/dataframe_data/hacrp_df.pkl')
main_df['HAC: Fiscal Year'] = main_df['HAC: Fiscal Year'].astype(str)
main_df.tail()
```

```python

print('main_df.shape:', main_df.shape)

ls = ['unplanned_visits_df.pkl', 'hvbp_clinical_outcomes_df.pkl', 'hvbp_efficiency_df.pkl',
      'hvbp_safety_df.pkl', 'hvbp_total_performance_scores_df.pkl', 'hvbp_hcahps_df.pkl',
      'Outpatient_Imaging_Efficiency_df.pkl', 'Payment_and_ValueOfCare_df.pkl', 
      'complication_and_deaths_df.pkl', 'Timely_and_Effective_Care_df.pkl', 
      'hospital_info_df.pkl', 'hcahps_df.pkl', 'hai_df.pkl', 'hrrp_df.pkl']

brk = 0
for l in ls:
    print(l)
    tdf = pd.read_pickle(main_dir + 'MergeFacilityFiles/dataframe_data/' + l)
    print('tdf shape:', tdf.shape)
    
    main_df = main_df.merge(tdf, how='outer', on=['Facility ID', 'file_month', 'file_year'])
    print('main_df.shape:', main_df.shape, '\n')

    for i in list(main_df):
        if '_x' in i:
            print('redundant column name:', i)
            brk = 1
            break
            
    if brk == 1:
        break

for n in ['Facility ID', 'file_month', 'file_year']:
    main_df.rename(columns={n: n + ': ' + n}, inplace=True)

del tdf
```

```python
ls = ['General Info: Facility Name',
      'General Info: Address',
      'General Info: City',
      'General Info: County',
      'General Info: State',
      'General Info: ZIP Code',
      'General Info: Effectiveness of Care National Comparison',
      'General Info: Efficient Use of Medical Imaging National Comparison', 
      'General Info: Emergency Services',
      'General Info: Hospital Ownership', 
      'General Info: Hospital Type',
      'General Info: Hospital overall rating', 
      'General Info: Meets Criteria for Meaningful Use of EHRs', 
      'General Info: Meets criteria for promoting interoperability of EHRs',
      'General Info: Mortality National Comparison',
      'General Info: Patient Experience National Comparison', 
      'General Info: Readmission National Comparison', 
      'General Info: Safety of Care National Comparison',
      'General Info: Timeliness of Care National Comparison',
     ]

ls.reverse()
ls.extend(['file_month: file_month', 'file_year: file_year', 'Facility ID: Facility ID'])

main_df['file date: file date'] = main_df['file_month: file_month'].astype(str) + '-' + main_df['file_year: file_year'].astype(str)
main_df['file date: file date'] = pd.to_datetime(main_df['file date: file date'], format='%m-%Y')

main_df['Name and Num: Name and Num'] = main_df['General Info: Facility Name'] + ' (' + main_df['Facility ID: Facility ID'].astype(str) + ')'
ls.extend(['file date: file date', 'Name and Num: Name and Num'])

for l in ls:
    col = main_df.pop(l)
    main_df.insert(0, col.name, col)

main_df.head()
```

```python
print(main_df['General Info: Hospital Ownership'].unique().tolist())

d = {
    'Government - Hospital District or Authority': 'Government', 
    'Government - State': 'Government',  
    'Government - Local': 'Government',  
    'Government - Federal': 'Government',  
    'Government-Federal': 'Government', 
    'Government': 'Government',  
    'Government Federal': 'Government', 
    'Government-State': 'Government', 
    'Department of Defense': 'Government', 
    
    'Proprietary': 'Proprietary for-profit',  
    np.nan: 'Not given',
    'Voluntary non-profit - Private': 'Voluntary non-profit',  
    'Voluntary non-profit-Private': 'Voluntary non-profit',   
    'Voluntary non-profit - Other': 'Voluntary non-profit',  
    'Voluntary non-profit-Other': 'Voluntary non-profit',  
    'Voluntary non-profit - Church': 'Voluntary non-profit', 
}

main_df['General Info: Hospital Ownership'].replace(to_replace=d, inplace=True)
print(main_df['General Info: Hospital Ownership'].unique().tolist())
```

```python
print(main_df['General Info: Hospital Type'].unique().tolist())

d = {
     'Acute Care Hospitals': 'Acute Care', 
     np.nan: 'Not given',  
     'Critical Access Hospitals': 'Critical Access', 
     #'Childrens', 
     'Acute Care - Department of Defense': 'Acute Care',  
     'ACUTE CARE - VETERANS ADMINISTRATION': 'Acute Care',  
     #'Psychiatric',
}

main_df['General Info: Hospital Type'].replace(to_replace=d, inplace=True)
print(main_df['General Info: Hospital Type'].unique().tolist())
```

```python
main_df.columns = [col.replace(',', ' ') for col in main_df.columns]  
```

```python
ls = list(main_df)

h1 = []
h2 = []

for l in ls:
    l1 = l[:l.index(":")]
    l2 = l[l.index(":")+2:]
    
    h1.append(l1)
    h2.append(l2)    
    
print(len(h1))
print(len(h2))
print(main_df.shape)
```

```python
tdf = pd.DataFrame(h2) 

tdf = tdf.T

tdf.columns = ls

tdf = pd.concat([tdf, main_df], ignore_index=True)

tdf.columns = [tdf.columns, h1]

tdf = tdf.droplevel(0, axis=1)

tdf.columns = pd.MultiIndex.from_arrays([tdf.columns, tdf.iloc[0].values])

tdf = tdf.iloc[1:]

main_df = tdf.copy(deep=True)
del tdf

main_df.head()
```

```python
hospitals = sorted(main_df[('Facility ID', 'Facility ID')].unique())
print(hospitals[0:4])
```

```python
report_categories = list(set(main_df.columns.get_level_values(0).tolist()))
print(report_categories)
```

```python
report_categories = list(set(main_df.columns.get_level_values(0).tolist()))
report_categories = [x for x in report_categories if str(x) != 'file_month']
report_categories = [x for x in report_categories if str(x) != 'file_year']
report_categories = [x for x in report_categories if str(x) != 'General Info']
report_categories = [x for x in report_categories if str(x) != 'Facility ID']
report_categories.sort()

with open('dataframe_data/Files_For_App/report_categories.csv', 'w+', newline='') as OUT:
    writer = csv.writer(OUT)
    writer.writerow(report_categories)
    
```

```python
s1 = main_df.columns.get_level_values(0).tolist()
s2 = main_df.columns.get_level_values(1).tolist()

sub_categories = []
for i, s in enumerate(s1):
    
    if s in ['file_month', 'file_year', 'General Info', 'Facility ID']:
        continue
    else:
        sub_categories.append(s + ', ' + s2[i])

with open('dataframe_data/Files_For_App/sub_categories.csv', 'w+', newline='') as OUT:
    writer = csv.writer(OUT)
    writer.writerow(sub_categories)
```
